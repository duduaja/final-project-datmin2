```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_csv('labelled_final.csv')
df
```

```python
df_dinormalisasi = df[['x', 'y']]
df_dinormalisasi
```

```python
#normalisasi
scaler = MinMaxScaler()

normalized = scaler.fit_transform(df_dinormalisasi)
```

```python
df_normal = pd.DataFrame(normalized, columns=['x', 'y'])
df_normal
```

```python
df['x'] = df_normal['x']
df['y'] = df_normal['y']
df
```

```python
df_normal.to_csv('df_normal.csv')
```
